# Classifiers

### Make necessary imports & definitions

```python
import os
import numpy as np
from collections import Counter
import time
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import learning_curve
from sklearn.ensemble import RandomForestRegressor
import graphviz
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split as tts
from sklearn.naive_bayes import GaussianNB

MAX_WORDS = 3000
```

### Define fetures and labels from the given data

```python
def make_dictionary(path):

    files = os.listdir(path)
    tweeter_feed_list = [path + tweeter_feed for tweeter_feed in files]
    words = []

    for tweeter_feed in tweeter_feed_list:
        f = open(tweeter_feed)
        blob = f.read()
        words += blob.split(" ")

    for i in range(len(words)):
        words[i] = words[i].lower()
        if not words[i].isalpha():
            words[i] = ""
    
    dictionary = Counter(words)
    print '\n\n'
    del dictionary[""]

    return dictionary.most_common(MAX_WORDS)   
```

```python
def extract_features(path, data_set):
    
    files = os.listdir(path)
    tweeter_feed_list = [path + tweeter_feed for tweeter_feed in files]
    
    feature_set = []
    labels = []
    
    for tweeter_feed in tweeter_feed_list:
        data = []
        f = open(tweeter_feed)
        words = f.read().split(" ")
        for entry in dictionary:
            # entry[0] -> key // entry[1] -> value
            data.append(words.count(entry[0]))
        feature_set.append(data)
        
        if data_set == 'Mail':
            # ham -> 0 // spam -> 1
            if "ham" in tweeter_feed:
                labels.append(0)
            elif "spam" in tweeter_feed:
                labels.append(1)
                
        elif data_set == 'Political Affiliation':
            # dem -> 0 // rep -> 1
            if "dem" in tweeter_feed:
                labels.append(0)
            elif "rep" in tweeter_feed:
                labels.append(1)

    return feature_set, labels
```

```python
def learning_curves(model, features, labels, train_sizes):
    
    train_sizes, train_scores, validation_scores = learning_curve(
                                                 model, features, labels,
                                                 train_sizes = train_sizes, cv = 5,
                                                 scoring = 'neg_mean_squared_error')
    
    train_scores_mean = -train_scores.mean(axis = 1)
    validation_scores_mean = -validation_scores.mean(axis = 1)
    
    plt.plot(train_sizes, train_scores_mean, label = 'Training error')
    plt.plot(train_sizes, validation_scores_mean, label = 'Validation error')

    plt.ylabel('MSE', fontsize = 14)
    plt.xlabel('Training set size', fontsize = 14)
    
    return plt
```

## Choose which data set you wish to analyse

```python
# Uncomment one of the following blocks:

# path = "../emails/"
# class_labels = ['ham', 'spam']
# data_set = 'Mail'

path = "../tweets/"
class_labels = ['dem', 'rep']
data_set = 'Political Affiliation'


# make dictionary from train data
dictionary = make_dictionary(path)

# estract features & labels from train data using the dictionary
features, labels = extract_features(path, data_set)

# slip train data from test data
train_features_matrix, test_features_matrix, train_labels, test_labels = tts(features, labels, test_size = 0.2)
```

### Decision Tree Classifier

```python
from sklearn import tree

dictionary_names = []
for entry in dictionary:
    dictionary_names.append(entry[0])
    
max_depth = [3 , 10]
accuracy = []
run_time = []

for d in max_depth:

    start = time.time()
    
    clf = tree.DecisionTreeClassifier(criterion = "gini", max_depth = d)
    clf.fit(train_features_matrix, train_labels)
    predicted_labels = clf.predict(test_features_matrix)
    
    end = time.time()
    run_time.append(end - start)

    print "Pre-pruning: max_depth = " + str(d)
    print "Accuracy score : " + '%.4f' % accuracy_score(test_labels, predicted_labels)
    print ""
    
    accuracy.append((float) ('%.4f' % accuracy_score(test_labels, predicted_labels)))

    dot_data = tree.export_graphviz(clf, feature_names = dictionary_names, class_names = class_labels) 
    graph = graphviz.Source(dot_data)
    graph.format = 'png'
    graph.render('../viz/Decision_Tree_' + str(data_set) + '.max_depth=' + str(d))

annotation = []
for i in max_depth:
    annotation.append('max_depth = ' + str(i))

plt.plot(run_time, accuracy, 'b')

for i, txt in enumerate(annotation):
    plt.annotate(txt, (run_time[i], accuracy[i]))

plt.xlabel('time (s)')
plt.ylabel('accuracy')
plt.title('Decision Tree ' + str(data_set) + ' Classifier')
plt.grid(True)
plt.savefig('../viz/Decision_Tree_plot_' + str(data_set) + '.png')
plt.show()

train_sizes = []
for i in range(1, 10, 2):
    train_sizes.append((int)((len(train_labels) * (float(i) / 10))))

plt.figure(figsize = (16, 5))
plt = learning_curves(clf, features, labels, train_sizes)
title = 'Learning curves for a ' + str(clf).split('(')[0] + ' model'
plt.title(title, fontsize = 18, y = 1.03)
plt.legend()
plt.savefig('../viz/Decision_Tree_learning_curve_' + str(data_set) + '.png')
plt.show()
```
